```python
import pandas as pd

# Read data
ds = pd.read_csv("amazon.csv")

# Convert to numeric FIRST
ds["rating"] = pd.to_numeric(ds["rating"], errors="coerce")
ds["rating_count"] = pd.to_numeric(ds["rating_count"], errors="coerce")

# Handle missing values (NO inplace)
ds["rating"] = ds["rating"].fillna(ds["rating"].median())
ds["rating_count"] = ds["rating_count"].fillna(0)

# Remove duplicates
ds = ds.drop_duplicates()

# Feature engineering
ds["popular"] = (ds["rating_count"] > ds["rating_count"].median()).astype(int)
ds["high_rating"] = (ds["rating"] >= 4).astype(int)
ds["engagement_score"] = ds["rating"] * ds["rating_count"]
ds["category_encoded"] = ds["category"].astype("category").cat.codes

# CLEAN AGAIN (NO inplace)
ds["engagement_score"] = ds["engagement_score"].fillna(0)

# FINAL CHECK
ds.isnull().sum()
```

```python
X = ds[[
    "rating_count",
    "popular",
    "engagement_score",
    "category_encoded"
]]

Y = ds["high_rating"]
```

```python
X=ds[["rating_count","popular","engagement_score","category_encoded"]]
Y=ds["high_rating"]
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(
    X ,Y, test_size=0.3,random_state=42
)
```

```python
model =LogisticRegression(max_iter=1000)
model.fit(X_train,Y_train)
Y_pred=model.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy)
```

```python
cm = confusion_matrix(Y_test, Y_pred)
print("Confusion Matrix:\n", cm)
```

```python
print(classification_report(Y_test, Y_pred))
```
